## Ejemplo 1: Estilos

### 2. Desarrollo:

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path1 = "https://raw.githubusercontent.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020/master/"
df = pd.read_csv(path1 + 'Datasets/melbourne_housing-clean.csv', index_col=0)

df.head()
```

Vamos a crear un histograma con los estilos default:

```python
sns.displot(df['price'], kind = 'hist', bins= 50);
```

Seaborn y matplotib incluyen algunos estilos predefinidos que podemos usar para cambiar el estilo visual de nuestras gr√°ficas:

darkgrid, whitegrid, dark, white, ticks


```python
sns.set_style('white')

sns.displot(df['price'], kind = 'hist', bins= 50);
```

```python
sns.set_style('whitegrid')

sns.displot(df['price'], kind = 'hist', bins= 50);
```

```python
sns.set_style('darkgrid')

sns.displot(df['price'], kind = 'hist', bins= 50);
```

```python
plt.style.use('fivethirtyeight')
sns.displot(df['price'], kind = 'hist', bins= 50);
```

```python
plt.style.use('ggplot')

sns.displot(df['price'], kind = 'hist', bins= 50);
```

```python
plt.style.use('classic')

sns.displot(df['price'], kind = 'hist', bins= 50);
```

Tambi√©n podemos usar estilos de manera 'temporal'. Esto quiere decir que el estilo s√≥lo aplica a las gr√°ficas que realicemos dentro de la sentencia `with`:

```python
plt.style.use('default')

with sns.axes_style('ticks'):
    sns.displot(df['price'], kind = 'hist');
```

```python
with plt.style.context('seaborn-bright'):
    sns.displot(df['price'], kind = 'hist', bins= 50);
```

Hay veces que tenemos que modificar nuestros estilos manualmente para obtener algo altamente personalizado, pero vale la pena echarle un ojo a los estilos predefinidos.
