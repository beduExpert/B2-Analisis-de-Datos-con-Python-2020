## Ejemplo 5: Gráficas de densidad

### 1. Objetivos:
    - Aprender a generar gráficas de densidad usando Seaborn.
    - Aprender a usar gráficas de densidad para comparar distribuciones de datos.

---
    
### 2. Desarrollo:

Podemos obtener gráficas de densidad usando el método `displot` de Seaborn,
sólo cambiando los argumentos que le pasamos:

```python
import numpy as np
import seaborn as sns
import pandas as pd
sns.set(style='whitegrid')
```

```python
laplace = np.random.laplace(loc=0.0, scale=1, size=10000)
sns.displot(laplace, kind= 'kde');
```

```python
chisquare = np.random.chisquare(4, size=10000)
sns.displot(chisquare, kind= 'kde');
```

Las gráficas de densidad son muy útiles para graficar varias distribuciones en la misma gráfica y poderlas comparar:

```python
df1 = pd.DataFrame({'grupo' : "normal1", 'datos': np.random.normal(loc=-2, scale=3, size=10000)})
df2 = pd.DataFrame({'grupo' : "normal2", 'datos': np.random.normal(loc=4.5, scale=1, size=10000)})
df3 = pd.DataFrame({'grupo' : "exponencial", 'datos': np.random.exponential(scale=1.0, size=10000) - 1})

df_concat = pd.concat([df1, df2, df3])
```

```python
sns.set(style='white')

sns.displot(data=df_concat, x="datos", hue="grupo", kind="kde", fill = True, alpha = 0.2)
```

```python
sns.kdeplot(data=df_concat, x="datos", hue="grupo",fill=True, alpha=.5)
```

Como puedes ver, usando gráficas de densidad obtenemos mucha claridad en la comparación. En este ejemplo agregamos `fill` y `alpha` para obtener el relleno y la transparencia. En una sesión posterior aprenderemos cómo estilizar nuestras gráficas a profundidad.
