# Ejemplo 5: Boxplots y Violinplots

1. Objetivos:
    - Aprender a generar boxplots y violinplots para analizar distribuciones de valores numéricos agrupados usando una variable categórica.

---
    
2. Desarrollo:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('white')
```

```python
path1 = 'https://github.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020/raw/master/Datasets/'
```

```python
df = pd.read_csv(path1 + '/athlete_olympic_events-clean.csv', index_col=0)

df.head()
```

```python
by_athlete = df.groupby(level=0)[['age', 'height', 'weight']].mean()
sex = df.groupby(level=0)['sex'].last()
merged = by_athlete.merge(sex, left_index=True, right_index=True)
```

```python
merged
```

La tabla `merged` tiene 3 variables numéricas y 1 variable categórica. Ya hemos visto cómo podemos analizar la distribución de una variable numérica utilizando boxplots:

```python
sns.boxplot(x = df['weight'])
```

Pero dado que tenemos una variable categórica (en este caso binaria, pues tiene solo dos valores posibles), podemos utilizar dos boxplots paralelos para analizar la distribución de esta misma variable numérica después de ser agrupada utilizando la variable categórica:

```python
plt.figure(figsize=(5, 10))
sns.boxplot(data=merged, x='sex', y='weight');
```

También podemos explorar las distribuciones usando un violinplot. Un violinplot es muy parecido a un boxplot, pero la diferencia es que en vez de graficar frecuencias grafica estimados de densidad. ¿Recuerdas las gráficas de densidad que utilizamos para hacer los histogramas más 'suaves' y poder comparar dos histogramas al mismo tiempo? Bueno, un violinplot es básicamente un boxplot mezclado con una gráfica de densidad.

En un violinplot es mucho más fácil ver ciertos detalles en la distribución de los datos que no es posible ver en los boxplots (ya que un boxplot no tiene mucha flexibilidad en cuanto a figuras se refiere). En cambio, en un violinplot es mucho más difícil ver los valores atípicos y dónde se encuentran ubicados.

```python
plt.figure(figsize=(5, 10))
sns.violinplot(data=merged, x='sex', y='weight');
```

La 'pildorita' negra en medio del violinplot representa el Rango Intercuartílico y el punto blanco la mediana. Así puedes comparar más fácilmente lo que estamos viendo en esta gráfica con la gráfica anterior.

```python
g = plt.figure(figsize=(5, 10))
g = sns.violinplot(data=merged, x='sex', y='weight', palette = 'crest')
g.set(title="Gráfica tipo violín")
```

## Grammar of graphics

Hay un modelo de cómo se programan las gráficas, este fue propuesto por Hadley Wickham, pueden leer el artículo aquí:
https://vita.had.co.nz/papers/layered-grammar.pdf

Inicialmente fue propuesto en R, pero se ha extendido a otros programas, incluyendo Python:
https://widdowquinn.github.io/Teaching-Data-Visualisation/exercises/grammar_of_graphics/grammar_of_graphics.html

(Seaborn 2022): This release debuts the seaborn.objects interface, an entirely new approach to making plots with seaborn. It is the product of several years of design and 16 months of implementation work. The interface aims to provide a more declarative, composable, and extensible API for making statistical graphics. It is **inspired by Wilkinson’s grammar of graphics**, offering a Pythonic API that is informed by the design of libraries such as ggplot2 and vega-lite along with lessons *from the past 10 years of seaborn’s development*.

https://seaborn.pydata.org/whatsnew/v0.12.0.html

While Seaborn doesn't explicitly adhere to a single, formal paradigm like ggplot's Grammar of Graphics, it does follow some core principles:

1. Layer-based design: Seaborn builds plots by adding layers of information on top of each other. This allows for flexibility and customization, letting you add elements like data points, lines, error bars, and annotations in a modular way.

2. Statistical data visualization: Seaborn is primarily focused on visualizing statistical data and relationships between variables. It provides a wide range of plot types specifically designed for this purpose, including histograms, scatter plots, box plots, heatmaps, and violin plots.

3. Aesthetic mapping: Seaborn uses a concept called "aesthetic mapping" to associate data attributes with visual properties like color, size, and shape. This allows you to visually encode information in your data without cluttering the plot with text labels.

4. Faceting and subplots: Seaborn allows you to create faceted plots and subplots, which can be helpful for visualizing data across different conditions or categories. This can be achieved through the facet_grid function or the subplots function.

5. Theme system: Seaborn has a built-in theme system that allows you to customize the overall look and feel of your plots. You can change things like the color palette, font size, and line styles to create plots that are consistent with your branding or project requirements.

6. Integration with Matplotlib: Seaborn is built on top of Matplotlib, a popular Python library for creating graphs. This allows Seaborn to leverage the extensive functionality of Matplotlib while providing a more user-friendly and statistically-oriented interface.
